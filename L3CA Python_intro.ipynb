# INTRODUCTION TO PYTHON

## 1. PROGRAMMING LANGUAGE

**Programming language** is a formal language designed to express instructions that can be executed by a computer or other computing device (Moble,Chipset Arduino, IoT,...). It consists of a set of rules, symbols, and instructions that are used to create software programs, applications, and computer systems.Programming languages provide a way for humans to communicate with machines in a way that is both structured and logical. 

## 2. ALGORITHM

An **algorithm** is a set of well-defined instructions or rules that can be followed to solve a problem or complete a task. It is a step-by-step procedure that can be executed by a computer or by a human to solve a particular problem or to perform a specific task.These instructions are written in a programming language that a computer can understand and execute.

## 3. COMPILATION

**Compilation** refers to the process of translating **human-readable code** written in a high-level programming language into **machine-readable object code** (binary code) that can be executed by a computer, in a single pass. This process is typically carried out by a **compiler**, which is a software program that takes the source code as input and generates the corresponding object code as output.

During the compilation process, the compiler analyzes the source code to check for syntax errors, type errors, and other issues. It then translates the code into a form that the computer can understand and execute.

## 4. INTERPRETATION

Code **interpretation** involves analyzing and executing code **line-by-line** in real-time, as opposed to translating the entire code into machine-executable code in a single pass via the compilation process. The interpreter reads each line of code, analyzes it, and executes it immediately, which can result in **slower performance** compared to compiled code. Interpretation is often used in scripting languages, such as Python, Ruby, and JavaScript, which are designed for rapid development and prototyping. 

## 5. CLASSES OF PROGRAMMING LANGUAGES

Programming languages can be classified in different ways based on various criteria, such as :
- **High-level** vs **Middle level** vs **low-level** Programming languages
- Procedural vs Object-oriented vs Functional Programming languages
- Compiled vs Interpreted Programming languages

### A. LOW LEVEL LANGUAGE

**Low-level language** refers to programming languages that are **closer to the machine code and hardware architecture** of a computer. These languages are typically more **difficult** to read and write than high-level languages, but they offer greater control over the computer's resources and can be **more efficient in terms of speed and memory usage**.

Examples of low-level languages include **assembly language**, which is a symbolic representation of machine code instructions, and **machine language**, which is the binary code that the computer can directly execute. These languages are often used for system-level programming tasks such as writing operating systems, device drivers, and embedded systems.

### B. HIGH LEVEL LANGUAGE

**High-level** language refers to programming languages that are designed to be **easier to read, write, and understand** than low-level languages. These languages provide more powerful features and constructs that allow developers to express complex ideas and algorithms more easily.

Examples of high-level languages include **Python**, Fortran, Java, and JavaScript. These languages typically offer features such as object-oriented programming, dynamic memory allocation and extensive libraries that provide a wide range of functionality.


### C. MIDDLE LEVEL LANGUAGE 

**Middle-level** languages are programming languages that **combine features of both high-level and low-level languages**. They provide a balance between the **ease of use and expressiveness** of high-level languages and the **performance and control** of low-level languages.

The only middle-level languages are **C/C++**. These languages offer features such as pointers, **memory management**, and **direct access to hardware resources**, but also provide abstractions and libraries that make it easier to write complex programs.

C/C++ are often used for system-level programming tasks such as writing device drivers, operating systems, and network protocols. They are also used for performance-critical applications such as game development, scientific computing, and image processing.

4. Calculate the **rate constant k** of a chemical reaction in function of **activation energy** J/mol and **temperature** in K,  the gas constant **R = 8.314** and Arrhenius equation is :  **k = exp(-E_activation/(R*temp))**

```python
import math

# Enter the activation energy and temperature
E_activation = float(input("Enter the activation energy (J/mol): "))
temperature = float(input("Enter the temperature (K): "))
R = 8.314

# Calculate the rate constant
if temperature > 0:
    k = math.exp(-E_activation / (R * temperature))
    print("Rate constant (k) =", k)

else:
    print("Invalid temperature. Temperature must be greater than 0 K.")
```

5. Calculate the **Gibbs free energy change** of a chemical reaction based on its **standard enthalpy change** in kJ/mol, ***entropy change** in J/(mol K) and **temperature** in K: **delta_G = delta_H - temperature * delta_S / 1000**

```python
delta_H = float(input("Enter the standard enthalpy change (kJ/mol): "))
delta_S = float(input("Enter the entropy change (J/(mol K)): "))
temperature = float(input("Enter the temperature (K): "))

if temperature > 0:
    delta_G = delta_H - (temperature * delta_S / 1000)
    print("Gibbs free energy change (delta_G) =", delta_G, "kJ/mol")
else:
    print("Invalid temperature. Temperature must be greater than 0 K.")
```

5. Calculate the **boiling point** of a liquid based on its vapor pressure: **boiling_point = -log(vapor_pressure)*R*298/(40.7-(-math.log(vapor_pressure)*R))** so that : R = 0.082, vapor_pressure = [0.1, 0.2, 0.05, 0.01, -0.3, 0]

```python
import math

R = 0.082
vapor_pressure = [0.1, 0.2, 0.05, 0.01, -0.3, 0]

for pressure in vapor_pressure:
    if pressure <= 0:
        print("Invalid vapor pressure:", pressure)
    else:
        boiling_point = (-math.log(pressure) * R * 298) / (40.7 - (-math.log(pressure) * R))
        print("Boiling point for vapor pressure", pressure, "is", boiling_point) 
```

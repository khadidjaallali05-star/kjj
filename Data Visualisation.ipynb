# Data visualisation with matplotlib


Data visualization with Matplotlib is a powerful way to create static, interactive, and animated visualizations in Python. Here's a quick guide to get started with common types of visualizations:

# 1. Basic Line Plot

A line plot is a type of chart that displays data points connected by a continuous line, typically used to represent trends or changes in data over a range. It is often used for:

    Tracking trends: Showing how variables change over time.
    Comparing datasets: Visualizing multiple data series for comparison.
    Highlighting relationships: Identifying patterns or correlations between variables.

Key Components of a Line Plot:

    X-axis (Horizontal): Represents the independent variable (e.g., time, categories).
    Y-axis (Vertical): Represents the dependent variable (e.g., values, measurements).
    Line: Connects the data points to indicate progression or trends.
    Data points: Markers (optional) representing specific values in the dataset.

Example Use Cases:

    Stock price trends over days.
    Temperature variations throughout the year.
    Population growth over decades.

```python
# Example 

import matplotlib.pyplot as plt
import numpy as np

# Data Example 
x = np.linspace(0, 10, 101)  # Generate 100 points between 0 and 10
y = np.sin(x)                # Sine f unction values
#print(y)

# Create the line plot
plt.plot(x, y, label="Sine Wave", color="red", linestyle="--", linewidth=2)

# Add labels and title
plt.title("Line Plot Example", fontsize=16)
plt.xlabel("X-axis (radians)", fontsize=12)
plt.ylabel("Y-axis (sin(x))", fontsize=12)

# Add a legend
plt.legend()

# Add grid lines
plt.grid(True, linestyle="--", alpha=0.7)


#plt.xlim(0, 10)
#plt.ylim(0,1)
# Show the plot
plt.show()
```

Explanation Features of Code

    Data:
        x: Shared values for the independent variable.
        y1, y2, y3: Three dependent datasets with different mathematical functions.

    Line Properties:
        Color: Distinguishes each line (blue, red, green).
        Line Style: Differentiates lines visually (--, -, -.).
        Line Width: Adjusts line thickness for emphasis.

    Legend:
        The plt.legend() function adds labels for each line.
        The loc parameter specifies the location of the legend (upper right).

    Grid:
        Enhances readability with dashed grid lines.

    Axis Limits:
        Controlled with plt.xlim() and plt.ylim() for a focused view.



# Multiple Lines in One Plot

A multiple-line plot is a type of line plot that displays two or more datasets on the same axes, allowing for easy comparison between them. Here's an example and explanation:


```python
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot multiple lines
plt.plot(x, y1, label="Sine Wave", color="blue", linestyle="--")
plt.plot(x, y2, label="Cosine Wave", color="red", linestyle=":")

# Add labels, title, legend, and grid
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)  # X-axis values
y1 = np.sin(x)               # First dataset: Sine wave
y2 = np.cos(x)               # Second dataset: Cosine wave
y3 = np.tan(x) / 10          # Third dataset: Scaled tangent wave

# Plot multiple lines
plt.plot(x, y1, label="Sine Wave", color="blue", linestyle="--", linewidth=2)
plt.plot(x, y2, label="Cosine Wave", color="red", linestyle="-.", linewidth=2)
plt.plot(x, y3, label="Tangent Wave (scaled)", color="green", linestyle="-", linewidth=1)

# Add title and labels
plt.title("Multiple Lines in One Plot", fontsize=16)
plt.xlabel("X-axis (radians)", fontsize=12)
plt.ylabel("Y-axis (function values)", fontsize=12)

# Customize grid
plt.grid(True, linestyle="--", alpha=0.5)

# Add a legend
plt.legend(loc="upper right", fontsize=10)

# Limit the axis range for better visibility
plt.xlim(0, 10)
plt.ylim(-2, 2)

# Show the plot
plt.show()
```

### Key Parameters in plt.subplots()

    nrows: Number of rows of subplots.
    ncols: Number of columns of subplots.
    figsize: Tuple (width, height) for overall figure size.
    sharex/sharey: If True, all subplots share the same X or Y axis.

```python
# Exemple 1
# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, axs = plt.subplots(2, 1)

# First subplot
axs[0].plot(x, y1, 'r', label="Sine")
axs[0].set_title("Sine Wave")
axs[0].legend()

# Second subplot
axs[1].plot(x, y2, 'b', label="Cosine")
axs[1].set_title("Cosine Wave")
axs[1].legend()

plt.tight_layout()
plt.show()
```
